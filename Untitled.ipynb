```python
import matplotlib.pyplot as plt
import numpy as np
    
```

```python
## OPERACIONES CON IMAGENES
```

```python
img = np.array(([1,2,4],[2,5,6],[8,9,5]))
```

```python
## Minimo y maximo
print(np.min(img), np.argmin(img))
print(np.max(img), np.argmax(img))
```

```python
## PROMEDIO
print(np.mean(img))
```

```python
## VARIANZA
print(np.var(img))
```

```python
## DESVIACION ESTANDAR
print(np.sqrt(np.var(img)))
```
